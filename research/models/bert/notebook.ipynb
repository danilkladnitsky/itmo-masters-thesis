```python
from transformers import pipeline

masked_pipeline = pipeline("mask-generation", model="bert-base-chinese")
result = masked_pipeline("我喜欢吃饭。")
print(result)
```

```python
from transformers import pipeline

fill_mask = pipeline("fill-mask", model="bert-base-chinese")
result = fill_mask("我在酒店喜欢喝[MASK]。")
print(result)
```

```python
from transformers import pipeline
corrector = pipeline("text2text-generation", model="blcuicall/gec-chinese")
result = corrector("他很高兴地跑。")  # Output: "他高兴地跑着。"
```
